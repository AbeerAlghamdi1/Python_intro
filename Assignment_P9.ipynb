### In each cell complete the task using pandas

```python
import pandas as pd
import numpy as np
```

Read in the titanic.csv file in the `~/data` directory as a pandas dataframe called **df**

```python
df = pd.read_csv('https://raw.githubusercontent.com/gumdropsteve/intro_to_python/main/day_09/data/titanic.csv')
df
```

Display the head of the dataframe

```python
df.head()
```

What is the percentage of people who survived? (hint find the mean of the survival column)

```python
survived_per = round(df.Survived.mean() * 100)
print(f'Survived Percentage is {survived_per}%')
```

How many women and how many men survived?

```python
#  SHOW ONLY SURVIVED FOR EACH FEMALE & MALE
female_survived = df.loc[(df.Survived == 1) & (df.Sex == 'female')].Survived.count()

male_survived = df.loc[(df.Survived == 1) & (df.Sex == 'male')].Survived.count()

print(f'Women Survived  {female_survived} \nMan Survived  {male_survived}')
```

What is the percentage of people that survied who paid a fare less than 10?

```python
survived_fare_people = df.loc[(df.Survived == 1) & (df.Fare < 10)]

survived_fare_per = survived_fare_people.Survived.count() / df.Survived.count() * 100
survived_fare_per

print(f'Survived & Fare < 10 Percentage is {survived_fare_per}%')
```

What is the average age of those who didn't survive?

```python
#df.loc[(df.Survived == 0)].Survived.count()

# df1 df for how did NOT Survived
df1 = df.loc[(df.Survived == 0)]
avg_age = round(df1.Age.mean())

print(f'Average age for didnt survive is {avg_age}')
```

What is the average age of those who did survive?

```python
# df2 df for how did  Survived
df2 = df.loc[(df.Survived == 1)]
avg_age = round(df2.Age.mean())

print(f'Average age for did survive is {avg_age}')
```

What is the average age of those who did and didn't survive grouped by gender?

```python
avg_age_not_survived = round(df1.groupby('Sex').Age.mean())
print(f'People didnt survive avg age based on {avg_age_not_survived}')

avg_age_survived = round(df2.groupby('Sex').Age.mean())
print(f'\nPeople survive avg age based on {avg_age_survived}')
```

## Tidy GDP

Manipulate the GDP.csv file and make it tidy, the result should be a pandas dataframe with the following columns:
* Country Name
* Country Code
* Year
* GDP

```python
tidy_GDP = pd.read_csv('https://raw.githubusercontent.com/gumdropsteve/intro_to_python/main/day_09/data/gdp.csv ')
```

```python
tidy_GDP.head()
```

```python
# delete extra columns using del OR
del tidy_GDP["Indicator Name"]
del tidy_GDP["Indicator Code"]
del tidy_GDP["Unnamed: 62"]
```

```python
# delete extra columns using drop 
tidy_GDP.drop(['Indicator Name','Indicator Code','Unnamed: 62'], inplace=True, axis=1)
```

```python
tidy_GDP.head()
```

```python
# Combine the columns to have year column 
tidy_GDP = tidy_GDP.melt(id_vars=["Country Name", "Country Code"],var_name=["Year"],value_name="GDP")
```
