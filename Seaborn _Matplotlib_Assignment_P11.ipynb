```
# Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```
# Load data and do some EDA
df = sns.load_dataset("tips")

df.head()
```

```
df['tip_percentage'] = df['tip'] / df['total_bill']*100

df.head()
```

####Total tips per days of week

```
sns.barplot(data=df, x="day", y="tip", estimator=np.sum)
sns.set_palette("Set2")
plt.title("Total tips per days of the week")
plt.figure(figsize=(15,5))
```

```
# Use seaborn to create a scatter plot with more details

plt.figure(figsize=(15,5))
sns.scatterplot( x="tip_percentage", y="day",
                hue="size",
                size="size",
                data=df,
                ),

plt.title("Distribution of tip percentage over week days");
```

####Line Plot


```
df.plot("day","tip_percentage")
plt.title('Tip percentage over week days')
plt.xlabel('day')
plt.ylabel('tip_percentage')
plt.show()
```

####Scatter Plot


```
df.plot.scatter("tip","day")
plt.title('Tips over week days')
plt.xlabel('day')
plt.ylabel('tip')
plt.show()
```

####Bar Plot

```
plt.bar(x = df["day"],height=df["tip"])
plt.title('Tips over week days')
plt.xlabel('day')
plt.ylabel('tip')
plt.show()
```
